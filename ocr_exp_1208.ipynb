```
import easyocr
```

```
reader = easyocr.Reader(['ch_sim','en']) # need to run only once to load model into memory
```

```
import urllib.request
```

```
urllib.request.urlretrieve('https://i1.wp.com/9to5google.com/wp-content/uploads/sites/4/2020/03/pixel_4a_blass_billboard_2.jpg?w=1500&quality=82&strip=all&ssl=1',filename='inputfile.jpg')
```

```
from IPython.display import Image
Image('inputfile.jpg')
```

```
result = reader.readtext('inputfile.jpg')
```

```
result
```

```
for i in range(len(result)):
  print(result[i][1])
```
